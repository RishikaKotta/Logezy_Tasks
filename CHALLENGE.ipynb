```python
import os
from PIL import Image
path = 'C:\\Users\\Rishika\\Downloads\\challenge\\challenge'
# Store the image file names in a list as long as they are jpgs
images = []
for f in os.listdir(path):
    if os.path.splitext(f)[-1] == '.jpg':
        images.append(os.path.join(path, f))    
images
names = [name for name in os.listdir(path) if (name.endswith(".jpg") or name.endswith(".JPG")) ]
wt = []
ht = []
avg_bright = []
for i in range(len(images)):
    image = Image.open(images[i])
               
    # Calculating height and width
    height, width = image.size
    ht.append(height)
    wt.append(width)  
    # Calculating average brightness
    greyscale_image = image.convert('L')
    histogram = greyscale_image.histogram()
    pixels = sum(histogram)
    brightness = scale = len(histogram)

    for index in range(0, scale):
        ratio = histogram[index] / pixels
        brightness += ratio * (-scale + index)\

    avg_brightness = brightness/scale
    avg_bright.append(avg_brightness)
  
import pandas as pd
column_names = ["Name","Height","Width","Avg Brightness"]
data = pd.DataFrame(index=range(len(images)), columns=col_names)
for i in range(len(images)):
    data.iloc[i,0] = names[i]
    data.iloc[i,1] = ht[i]
    data.iloc[i,2] = wt[i]
    data.iloc[i,3] = avg_bright[i]
```
